CREDIT CARD FRAUD DETECTION

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
train_data=pd.read_csv("/content/fraudTrain.csv")
train_data
```

```python
test_data=pd.read_csv("/content/fraudTest.csv")
test_data
```

CONCATENATION

```python
# Concate the both dataset
data=pd.concat([train_data,test_data])
data
```

```python
# Reset the index
data.reset_index(drop=True)
```

```python
# Drop irrelevent Columns
data.drop(columns=['Unnamed: 0','cc_num','first','last','unix_time','trans_num','street','city'],inplace=True)
```

ENCODING

```python
# Encode the categorical features
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
label_cols=['trans_date_trans_time', 'merchant', 'category','gender','state', 'job', 'dob']
for col in label_cols:
  data[col]=le.fit_transform(data[col])
```

```python
# Convert the integers to float
int_cols=['amt',  'zip', 'city_pop',  'is_fraud']
data[int_cols]=data[int_cols].astype(float)
```

```python
# Count of each class
class_counts = data['is_fraud'].value_counts()
print("Transaction Distribution:\n", class_counts)

# Percentage
fraud_percent = data['is_fraud'].value_counts(normalize=True) * 100
print("\nPercentage Distribution:\n", fraud_percent)
```

```python
data.dropna(subset=['is_fraud'], inplace=True)
```

SPLIT FEATURES AND TARGET

```python
x=data.drop('is_fraud',axis=1)
y=data['is_fraud']
```

```python
# Splitting train-test data
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, random_state=42)
```

LOGISTIC REGRESSION

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
